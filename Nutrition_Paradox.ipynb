```
import requests
import pprint
import pandas as pd

# obesity adults

url = "https://ghoapi.azureedge.net/api/NCD_BMI_30C"

obesity_adults = requests.get(url)

obesity_data1 = obesity_adults.json()

# pprint.pprint(obesity_data1)

df_obesityadults =  pd.DataFrame(obesity_data1['value'])
df_obesityadults["age_group"]="Adults"
df_obesityadults.head()
```

```
#obesity children

url = "https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C"

obesity_children = requests.get(url)

obesity_data2= obesity_children.json()

df_obesitychildren = pd.DataFrame(obesity_data2['value'])
df_obesitychildren["age_group"]="Children"
df_obesitychildren.head()
```

```
#obesity
df_obesity = pd.concat([df_obesityadults, df_obesitychildren])
df_obesity.head()
```

```
#malnutrition adults

url = "https://ghoapi.azureedge.net/api/NCD_BMI_18C"

malnutrition_adults = requests.get(url)

malnutrition_data1= malnutrition_adults.json()

df_malnutritionadults = pd.DataFrame(malnutrition_data1['value'])
df_malnutritionadults["age_group"]="Adults"
df_malnutritionadults.head()
```

```
#malnutrition children

url = "https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C"

malnutrition_children = requests.get(url)

malnutrition_data2 = malnutrition_children.json()

df_malnutritionchildren = pd.DataFrame(malnutrition_data2['value'])
df_malnutritionchildren["age_group"]="Children"
df_malnutritionchildren.head()
```

```
#malnutrition
df_malnutrition = pd.concat([df_malnutritionadults, df_malnutritionchildren])
df_malnutrition.head()
```

```
# filtering year from 2012 to 2022

df_obesity = df_obesity[

    (df_obesity['TimeDim'] >= 2012) & (df_obesity['TimeDim'] <= 2022)

]
df_malnutrition = df_malnutrition[

    (df_malnutrition['TimeDim'] >= 2012) & (df_malnutrition['TimeDim'] <= 2022)
]
```

```
columns_to_keep = [
    'ParentLocation',
    'Dim1',
    'TimeDim',
    'Low',
    'High',
    'NumericValue',
    'SpatialDim',
    'age_group'
]
df_obesity = df_obesity [columns_to_keep]
df_malnutrition = df_malnutrition[columns_to_keep]

print(df_obesity.head())
print(df_malnutrition.head())
```

```
# rename and replacing of columns

df_obesity = df_obesity.rename(columns={
    'ParentLocation': 'Region',
    'Dim1': 'Gender',
    'TimeDim': 'Year',
    'Low': 'LowerBound',
    'High': 'UpperBound',
    'NumericValue': 'Mean_Estimate',
    'SpatialDim': 'Country'
})

df_obesity['Gender'] = df_obesity['Gender'].replace({
    'SEX_FMLE':'Female',
    'SEX_MLE': 'Male',
    'SEX_BTSX':'Both',

})

print(df_obesity.head())

df_malnutrition = df_malnutrition.rename(columns={
    'ParentLocation': 'Region',
    'Dim1': 'Gender',
    'TimeDim': 'Year',
    'Low': 'LowerBound',
    'High': 'UpperBound',
    'NumericValue': 'Mean_Estimate',
    'SpatialDim': 'Country'
})

df_malnutrition['Gender'] = df_malnutrition['Gender'].replace({
    'SEX_FMLE': 'Female',
    'SEX_MLE': 'Male',
    'SEX_BTSX': 'Both',

})

print(df_malnutrition.head())
```

```
import pycountry
special_cases = {
                    'GLOBAL': 'Global',
                    'WB_LMI': 'Low & Middle Income',
                    'WB_HI': 'High Income',
                    'WB_LI': 'Low Income',
                    'EMR': 'Eastern Mediterranean Region',
                    'EUR': 'Europe',
                    'AFR': 'Africa',
                    'SEAR': 'South-East Asia Region',
                    'WPR': 'Western Pacific Region',
                    'AMR': 'Americas Region',
                    'WB_UMI': 'Upper Middle Income'}

def code_to_country(code):
    try:
        return pycountry.countries.get(alpha_3=code.upper()).name
    except:
        if code in special_cases:
            return special_cases[code]
        else:
            return code

df_obesity['Country'] = df_obesity ['Country'].apply(code_to_country)

df_malnutrition['Country'] = df_malnutrition['Country'].apply(code_to_country)

# print(df_obesity['Country'].unique())
# print(df_malnutrition['Country'].unique())
```

```
# CI_Width column creation
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']

df_obesity.head()

df_malnutrition ['CI_Width'] = df_malnutrition ['UpperBound'] - df_malnutrition['LowerBound']

df_malnutrition.head()
```

```
#obesity_level & malnutrition_level
def categorize_obesity(value):
    if value >= 30:
        return 'High'
    elif 25 <= value < 30:
        return 'Moderate'
    else:
        return 'Low'

df_obesity['obesity_level'] = df_obesity['Mean_Estimate'].apply(categorize_obesity)
df_obesity.head()

def categorize_malnutrition(value):
    if value >= 20:
        return 'High'
    elif 10 <= value < 20:
        return 'Moderate'
    else:
        return 'Low'

df_malnutrition['malnutrition_level'] = df_malnutrition['Mean_Estimate'].apply(categorize_malnutrition)
df_malnutrition.head()
```

```
# Exploratory Data Analysis (EDA) with Python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for Mean_Estimate
sns.histplot(df_obesity['Mean_Estimate'], kde=True)
plt.title('Distribution of Obesity Mean Estimate')
plt.show()

# CI Width
sns.histplot(df_obesity['CI_Width'], kde=True)
plt.title('Distribution of Confidence Interval Width (Obesity)')
plt.show()
```

9.Countries where female obesity exceeds male by large margin (same year)


```
df_q9 = pd.read_sql_query(''' SELECT
  o1.Country,
  o1.Year,
  o1.Mean_Estimate AS Female_Obesity,
  o2.Mean_Estimate AS Male_Obesity,
  (o1.Mean_Estimate - o2.Mean_Estimate) AS Difference
FROM
  obesity o1
JOIN
  obesity o2
ON
  o1.Country = o2.Country AND o1.Year = o2.Year
WHERE
  o1.Gender = 'Female' AND o2.Gender = 'Male'
  AND (o1.Mean_Estimate - o2.Mean_Estimate) > 5
Group by o1.Country
ORDER BY
  Difference DESC;
''',conn)

print(df_q9)
```

```
%%writefile app.py
import streamlit as st
import seaborn as sns
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

st.set_page_config(page_title="Nutrition Paradox Dashboard", layout="wide")

st.title("Nutrition Paradox: Obesity & Malnutrition Dashboard")

st.write("""
This dashboard shows the full process:
- Data Cleaning & Preparation
- 25 SQL Queries
- EDA Visualizations
- Combined Insights
- Final Recommendations
""")

conn = sqlite3.connect("nutrition.db", check_same_thread=False)


st.header("Data Overview")


df_obesity = pd.read_sql_query("SELECT * FROM obesity LIMIT 100", conn)
st.subheader("Obesity Table (Sample)")
st.write(df_obesity.head())

df_malnutrition = pd.read_sql_query("SELECT * FROM malnutrition LIMIT 100", conn)
st.subheader("Malnutrition Table (Sample)")
st.write(df_malnutrition.head())

st.header("QUERY")
tab1, tab2, tab3 = st.tabs(["Obesity Queries", "Malnutrition Queries", "Combined Queries"])


with tab1:

    # 1
    st.subheader("1.Top 5 regions with the highest average obesity levels in the most recent year(2022)")
    df_q1 = """SELECT Region, round(AVG(Mean_Estimate),2) AS Avg_Obesity
    FROM obesity
    WHERE Year = 2022
    GROUP BY Region
    ORDER BY Avg_Obesity DESC
    LIMIT 5;
    """
    df_q1 = pd.read_sql_query(df_q1, conn)
    df_q1.index += 1
    st.dataframe(df_q1)
    st.plotly_chart(px.bar(df_q1, x='Avg_Obesity', y='Region', orientation='h', title='Top 5 Regions - Obesity (2022)', color_continuous_scale = ["#ffffff", "#004080"], color='Avg_Obesity'))

    # 2
    st.subheader("2.Top 5 countries with highest obesity estimates")
    df_q2 = """SELECT Country, round(AVG(Mean_Estimate),2) AS Avg_Obesity
    FROM obesity
    GROUP BY Country
    ORDER BY Avg_Obesity DESC
    LIMIT 5;
    """
    df_q2 = pd.read_sql_query(df_q2, conn)
    df_q2.index += 1
    st.dataframe(df_q2)
    st.plotly_chart(px.bar(df_q2, x='Avg_Obesity', y='Country', orientation='h', title='Top 5 Countries - Obesity', color_continuous_scale = ["#ffffff", "#004080"], color='Avg_Obesity'))

    # 3
    st.subheader("3.Obesity trend in India over the years")
    df_q3 = """SELECT Year, round(AVG(Mean_Estimate),2) AS Avg_Obesity
    FROM obesity
    WHERE Country = 'India'
    GROUP BY Year
    ORDER BY Year;
    """
    df_q3 = pd.read_sql_query(df_q3, conn)
    df_q3.index += 1
    st.dataframe(df_q3)
    st.plotly_chart(px.line(df_q3, x='Year', y='Avg_Obesity', markers=True, title='Obesity Trend in India'))

    # 4
    st.subheader("4.Average obesity by gender")
    df_q4 = """SELECT Gender, round(AVG(Mean_Estimate),2) AS Avg_Obesity
    FROM obesity
    GROUP BY Gender;
    """
    df_q4 = pd.read_sql_query(df_q4, conn)
    df_q4.index += 1
    st.dataframe(df_q4)
    st.plotly_chart(px.bar(df_q4, x='Gender', y='Avg_Obesity', title='Obesity by Gender', color='Gender', text='Avg_Obesity'))

    # 5
    st.subheader("5.Country count by obesity level category and age group")
    df_q5 = """SELECT obesity_level, age_group, COUNT(DISTINCT Country) AS Country_Count
    FROM obesity
    GROUP BY obesity_level, age_group;
    """
    df_q5 = pd.read_sql_query(df_q5, conn)
    df_q5.index += 1
    st.dataframe(df_q5)
    st.plotly_chart(px.bar(df_q5, x='Age_Group', y='Country_Count', color='Obesity_Level', title='Country Count by Obesity Level & Age Group', barmode='group'))

    # 6
    st.subheader("6.Top 5 countries least reliable countries and Top 5 most consistent countries")
    df_q6a = """SELECT Country, round(AVG(CI_Width),2) AS Avg_CI_Width
    FROM obesity
    GROUP BY Country
    ORDER BY Avg_CI_Width DESC
    LIMIT 5;
    """
    df_q6a= pd.read_sql_query(df_q6a, conn)
    df_q6a.index += 1
    st.dataframe(df_q6a)
    st.plotly_chart(px.bar(df_q6a, x='Avg_CI_Width', y='Country', orientation='h', color_continuous_scale = ["#ffffff", "#004080"],title='Top 5 Least Reliable (Highest CI Width)', color='Avg_CI_Width'))


    df_q6b = """SELECT Country, round(AVG(CI_Width),2) AS Avg_CI_Width
    FROM obesity
    GROUP BY Country
    ORDER BY Avg_CI_Width ASC
    LIMIT 5;
    """
    df_q6b= pd.read_sql_query(df_q6b, conn)
    df_q6b.index += 1
    st.dataframe(df_q6b)
    st.plotly_chart(px.bar(df_q6b, x='Avg_CI_Width', y='Country', orientation='h',title='Top 5 Most Consistent (Lowest CI Width)', color_continuous_scale = ["#ffffff", "#004080"],color='Avg_CI_Width'))


    # 7
    st.subheader("7.Average obesity by age group")
    df_q7 = """SELECT age_group, round(AVG(Mean_Estimate),2) AS Avg_Obesity
    FROM obesity
    GROUP BY age_group;
    """
    df_q7 = pd.read_sql_query(df_q7, conn)
    df_q7.index += 1
    st.dataframe(df_q7)
    st.plotly_chart(px.bar(df_q7, x='Age_Group', y='Avg_Obesity', color='Age_Group', color_continuous_scale = ["#ffffff", "#004080"],title='Average Obesity by Age Group'))


    # 8
    st.subheader("8.Top 10 Countries with consistent low obesity over the years")
    df_q8 = """SELECT Country, round(AVG(Mean_Estimate),2) AS Avg_Obesity,round(AVG(CI_Width),2) AS Avg_CI_Width
    FROM obesity
    GROUP BY Country
    HAVING Avg_Obesity < 25 AND Avg_CI_Width < 2
    ORDER BY Avg_Obesity ASC
    LIMIT 10;
    """
    df_q8 = pd.read_sql_query(df_q8, conn)
    df_q8.index += 1
    st.dataframe(df_q8)

    st.plotly_chart(px.bar(df_q8, x='Avg_Obesity', y='Country', orientation='h',title='Consistently Low Obesity Countries', color_continuous_scale = ["#ffffff", "#004080"] ,color='Avg_Obesity'))

    # 9
    st.subheader("9.Countries where female obesity exceeds male by large margin")
    df_q9 = """ SELECT
        o1.Country,
        o1.Year,
        round(o1.Mean_Estimate,2) AS Female_Obesity,
        round(o2.Mean_Estimate,2) AS Male_Obesity,
        round((o1.Mean_Estimate - o2.Mean_Estimate),2) AS Difference
      FROM
        obesity o1
      JOIN
        obesity o2
      ON
        o1.Country = o2.Country AND o1.Year = o2.Year
      WHERE
        o1.Gender = 'Female' AND o2.Gender = 'Male'
        AND (o1.Mean_Estimate - o2.Mean_Estimate) > 5
      GROUP BY
        o1.Country
      ORDER BY
        Difference DESC;
     """
    df_q9 = pd.read_sql_query(df_q9, conn)
    df_q9.index += 1
    st.dataframe(df_q9)
    st.plotly_chart(px.scatter(df_q9, x='Year', y='Difference', color='Country',size='Difference', hover_name='Country', title='Female vs Male Obesity Gap > 5'))


    # 10
    st.subheader("10.Global average obesity percentage per year")
    df_q10 = """SELECT Year, round(AVG(Mean_Estimate),2) AS Global_Avg_Obesity
    FROM obesity
    GROUP BY Year
    ORDER BY Year;
    """
    df_q10 = pd.read_sql_query(df_q10, conn)
    df_q10.index += 1
    st.dataframe(df_q10)
    st.plotly_chart(px.line(df_q10, x='Year', y='Global_Avg_Obesity', markers=True,title='Global Average Obesity Over Time'))


with tab2:

    # 1
    st.subheader("1. Avg. malnutrition by age group")
    q1_df = pd.read_sql_query("""
        SELECT age_group, AVG(Mean_Estimate) AS Avg_Malnutrition
        FROM malnutrition
        GROUP BY age_group;
    """, conn)
    q1_df.index += 1
    st.dataframe(q1_df)
    st.plotly_chart(px.bar(q1_df, x='Age_Group', y='Avg_Malnutrition', title='Average Malnutrition by Age Group', color='Age_Group'))


    # 2
    st.subheader("2. Top 5 countries with highest malnutrition")
    q2_df = pd.read_sql_query("""
        SELECT Country, AVG(Mean_Estimate) AS Avg_Malnutrition
        FROM malnutrition
        GROUP BY Country
        ORDER BY Avg_Malnutrition DESC
        LIMIT 5;
    """, conn)
    q2_df.index += 1
    st.dataframe(q2_df)
    st.plotly_chart(px.bar(q2_df, x='Country', y='Avg_Malnutrition', color='Country', title='Top 5 Countries with Highest Malnutrition'))


    # 3
    st.subheader("3. Malnutrition trend in African region over the years")
    q3_df = pd.read_sql_query("""
        SELECT Year, AVG(Mean_Estimate) AS Avg_Malnutrition
        FROM malnutrition
        WHERE Region = 'Africa'
        GROUP BY Year
        ORDER BY Year;
    """, conn)
    q3_df.index += 1
    st.dataframe(q3_df)
    st.plotly_chart(px.line(q3_df, x='Year', y='Avg_Malnutrition', title='Malnutrition Trend in Africa'))


    # 4
    st.subheader("4. Gender-based average malnutrition")
    q4_df = pd.read_sql_query("""
        SELECT Gender, AVG(Mean_Estimate) AS Avg_Malnutrition
        FROM malnutrition
        GROUP BY Gender;
    """, conn)
    q4_df.index += 1
    st.dataframe(q4_df)
    st.plotly_chart(px.pie(q4_df, names='Gender', values='Avg_Malnutrition', title='Malnutrition by Gender'))


    # 5.
    st.subheader("5. Malnutrition level-wise (average CI_Width by age group)")
    df_q5 = pd.read_sql_query("""
        SELECT malnutrition_level, age_group, AVG(CI_Width) AS Avg_CI_Width
        FROM malnutrition
        GROUP BY malnutrition_level, age_group;
    """, conn)
    df_q5.index += 1
    st.dataframe(df_q5)
    st.plotly_chart(px.bar(df_q5, x='Age_Group', y='Avg_CI_Width', color='Malnutrition_Level', barmode='group', title='CI_Width by Malnutrition Level & Age Group'))


    # 6
    st.subheader("6. Yearly malnutrition change in specific countries")
    df_q6 = pd.read_sql_query("""
        SELECT Country, Year, AVG(Mean_Estimate) AS Avg_Malnutrition
        FROM malnutrition
        WHERE Country IN ('India', 'Nigeria', 'Brazil')
        GROUP BY Country, Year
        ORDER BY Country, Year;
    """, conn)
    df_q6.index += 1
    st.dataframe(df_q6)
    st.plotly_chart(px.line(df_q6, x='Year', y='Avg_Malnutrition', color='Country', title='Malnutrition Trend: India, Nigeria, Brazil'))


    # 7
    st.subheader("7. Regions with lowest malnutrition averages")
    df_q7 = pd.read_sql_query("""
        SELECT Region, AVG(Mean_Estimate) AS Avg_Malnutrition
        FROM malnutrition
        GROUP BY Region
        ORDER BY Avg_Malnutrition ASC;
    """, conn)
    df_q7.index += 1
    st.dataframe(df_q7)
    st.plotly_chart(px.bar(df_q7, x='Region', y='Avg_Malnutrition', title='Malnutrition Averages by Region'))


    # 8
    st.subheader("8. Countries with increasing malnutrition")
    df_q8 = pd.read_sql_query("""
        SELECT Country, MIN(Mean_Estimate) AS Min_Malnutrition,
              MAX(Mean_Estimate) AS Max_Malnutrition,
              (MAX(Mean_Estimate) - MIN(Mean_Estimate)) AS Difference
        FROM malnutrition
        GROUP BY Country
        HAVING Difference > 0
        ORDER BY Difference DESC;
    """, conn)
    df_q8.index += 1
    st.dataframe(df_q8)
    st.plotly_chart(px.bar(df_q8.head(10), x='Country', y='Difference', color='Country',title='Top Countries with Increasing Malnutrition'))


    # 9
    st.subheader("9. Min/Max malnutrition levels year-wise comparison")
    df_q9 = pd.read_sql_query("""
        SELECT Year, MIN(Mean_Estimate) AS Min_Malnutrition,
              MAX(Mean_Estimate) AS Max_Malnutrition
        FROM malnutrition
        GROUP BY Year
        ORDER BY Year;
    """, conn)
    df_q9.index += 1
    st.dataframe(df_q9)
    st.plotly_chart(px.line(df_q9, x='Year', y=['Min_Malnutrition', 'Max_Malnutrition'], title='Year-wise Min and Max Malnutrition'))


    # 10
    st.subheader("10. High CI_Width flags for monitoring")
    df_q10 = pd.read_sql_query("""
        SELECT Country, Year, CI_Width
        FROM malnutrition
        WHERE CI_Width > 5
        GROUP BY Country
        ORDER BY CI_Width DESC;
    """, conn)
    df_q10.index += 1
    st.dataframe(df_q10)
    st.plotly_chart(px.scatter(df_q10, x='Year', y='CI_Width', color='Country', size='CI_Width', title='High CI_Width Records'))


with tab3:

    # 1
    st.subheader("1.Obesity vs Malnutrition Comparison (5 Countries)")
    df_c1 = """
    SELECT o.Country, AVG(o.Mean_Estimate) AS Avg_Obesity,
           AVG(m.Mean_Estimate) AS Avg_Malnutrition
    FROM obesity o
    JOIN malnutrition m ON o.Country = m.Country
    WHERE o.Country IN ('India', 'Nigeria', 'Brazil', 'Europe', 'China')
    GROUP BY o.Country;
    """
    df_c1 = pd.read_sql_query(df_c1, conn)
    df_c1.index += 1
    st.dataframe(df_c1)
    fig1 = px.bar(df_c1.melt(id_vars="Country", var_name="Metric", value_name="Average"),
              x="Country", y="Average", color="Metric",
              barmode="group", title="Obesity vs Malnutrition in Selected Countries")
    st.plotly_chart(fig1)


    # 2
    st.subheader("2.Gender-based disparity in both obesity and malnutrition")
    df_c2 = """
    SELECT o.Gender,
    AVG(o.Mean_Estimate) AS Avg_Obesity,
    AVG(m.Mean_Estimate) AS Avg_Malnutrition
    FROM obesity o
    JOIN malnutrition m
    ON o.Gender = m.Gender AND o.Year = m.Year AND o.Country = m.Country
    GROUP BY o.Gender;
    """
    df_c2= pd.read_sql_query(df_c2, conn)
    df_c2.index += 1
    st.dataframe(df_c2)
    fig2 = px.bar(df_c2.melt(id_vars="Gender", var_name="Metric", value_name="Average"),
              x="Gender", y="Average", color="Metric",

              barmode="group", title="Gender-Based Obesity vs Malnutrition")
    st.plotly_chart(fig2)


    # 3
    st.subheader("3.Region-wise avg estimates side-by-side(Africa and America)")
    df_c3 ="""
    SELECT o.Region,
    AVG(o.Mean_Estimate) AS Avg_Obesity,
    AVG(m.Mean_Estimate) AS Avg_Malnutrition
    FROM obesity o
    JOIN malnutrition m
    ON o.Region = m.Region AND o.Country = m.Country
    WHERE o.Region IN ('Africa', 'Americas')
    GROUP BY o.Region;
    """
    df_c3= pd.read_sql_query(df_c3, conn)
    df_c3.index += 1
    st.dataframe(df_c3)
    fig3 = px.bar(df_c3.melt(id_vars="Region", var_name="Metric", value_name="Average"),
              x="Region", y="Average", color="Metric",
              barmode="group", title="Region-wise Obesity vs Malnutrition (Africa & Americas)")
    st.plotly_chart(fig3)


    # 4
    st.subheader("4.Countries with obesity up & malnutrition down")
    df_c4 ="""SELECT o.Country,
    (MAX(o.Mean_Estimate) - MIN(o.Mean_Estimate)) AS Obesity_Change,
    (MAX(m.Mean_Estimate) - MIN(m.Mean_Estimate)) AS Malnutrition_Change
    FROM obesity o
    JOIN malnutrition m
    ON o.Country = m.Country AND o.Year = m.Year
    GROUP BY o.Country
    HAVING Obesity_Change > 0 AND Malnutrition_Change < 0
    ORDER BY Obesity_Change DESC;
    """
    df_c4= pd.read_sql_query(df_c4, conn)
    df_c4.index += 1
    st.dataframe(df_c4)
    fig4 = px.bar(df_c4.melt(id_vars="Country", var_name="Metric", value_name="Change"),
              x="Country", y="Change", color="Metric",
              barmode="group", title="Obesity↑ & Malnutrition↓ Change by Country")
    st.plotly_chart(fig4)


    # 5
    st.subheader("5.Age-wise trend analysis")
    df_c5 = """
    SELECT o.age_group,
    AVG(o.Mean_Estimate) AS Avg_Obesity,
    AVG(m.Mean_Estimate) AS Avg_Malnutrition
    FROM obesity o
    JOIN malnutrition m
    ON o.age_group = m.age_group AND o.Country = m.Country AND o.Year = m.Year
    GROUP BY o.age_group;
    """
    df_c5= pd.read_sql_query(df_c5, conn)
    df_c5.index += 1
    st.dataframe(df_c5)
    fig5 = px.bar(df_c5.melt(id_vars="Age_Group", var_name="Metric", value_name="Average"),
              x="Age_Group", y="Average", color="Metric",
              barmode="group", title="Age-wise Obesity vs Malnutrition")
    st.plotly_chart(fig5)

```
